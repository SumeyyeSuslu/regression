```python
import os
file_paths = [] 
for root, dirs, files in os.walk("./DummyGeometry"):
    for file in files:
        if file.endswith(".plt"):
             file_paths.append(os.path.join(root, file))
print(file_paths)
```

```python
# Spark Session, Pipeline, Functions, and Metrics
from pyspark import SparkContext, SparkConf
from pyspark.sql import SQLContext, SparkSession
from pyspark.ml.feature import OneHotEncoder, StringIndexer, StandardScaler, VectorAssembler
from pyspark.ml import Pipeline, PipelineModel
from pyspark.sql.functions import rand
from pyspark.ml.stat import Correlation
from pyspark.ml.linalg import Vector
from pyspark.ml.feature import VectorIndexer
from pyspark.ml.evaluation import RegressionEvaluator
from sparktorch import serialize_torch_obj, SparkTorch, serialize_torch_obj_lazy
import torch
import torch.nn as nn

from sparktorch import PysparkPipelineWrapper
from pyspark.sql.types import *
from pyspark.sql.functions import col, udf, column, rand
from pyspark.mllib.evaluation import RegressionMetrics
```

```python
spark = SparkSession \
    .builder \
    .appName("Python Spark SQL basic example") \
    .getOrCreate()
```

```python
import re
import numpy as np
import pltreader as plt

dataset = []
for item in file_paths:
  file_content={}
  pat = re.compile("mach(.*)_aoa(.*)_beta(.*)/")
  obj = pat.search(item)
  mach = float(obj.group(1))
  aoa = float(obj.group(2))
  beta = float(obj.group(3))
  with open( item, 'rb') as f:
    bytelist = f.read()
  header = plt.read_header(bytelist)
  data = plt.read_data(bytelist,header)
  for zone_id in range (len(data['Zones'])):
    zone_shape = data['Zones'][zone_id]['Pressure'].shape
    data['Zones'][zone_id]['mach']=mach*np.ones(zone_shape, dtype = 'float32')
    data['Zones'][zone_id]['aoa']=aoa*np.ones(zone_shape, dtype = 'float32')
    data['Zones'][zone_id]['beta']=beta*np.ones(zone_shape, dtype = 'float32')


  file_content['data'] = data
  dataset.append(file_content)
```

```python
cols=['mach','aoa','beta']+header['VarNames']
zid=0
m=[]
for fid in range(len(file_paths)):
  t=[]
  for col in cols:
    t.append(dataset[fid]['data']['Zones'][zid][col])
  m.append(np.vstack(t))
s=np.hstack(m).transpose()
```

```python
# dataset[file_id]['data']['Zones'][zone_id][var_name])
```

```python
df=spark.createDataFrame(s.tolist(),cols)
```

```python
df.limit(5).toPandas()
```

```python
import plotly.express as px

fig = px.scatter_3d(df.toPandas(),x='CoordinateX', y='CoordinateY', z='CoordinateZ', color='Pressure')

fig.update_scenes(aspectmode='data')
fig.show()
```

```python
vector_assembler= VectorAssembler(
    inputCols=['mach','aoa','CoordinateX', 'CoordinateY','CoordinateZ' ],outputCol='features')
```

```python
# Define network

from MLPpressure import MLP
network = MLP()
```

```python
# Build the pytorch object
torch_obj = serialize_torch_obj(
    model=network,
    criterion=nn.L1Loss(),
    optimizer=torch.optim.Adam,
    lr=0.01)
```

```python
spark_model = SparkTorch(
        inputCol='features',
        labelCol='Pressure',
        predictionCol='predictions',
        torchObj=torch_obj,
        iters=2000,
        verbose=1,
        miniBatch=8,
        earlyStopPatience=40,
        validationPct=0.2
    )
```

```python
# Create and save the Pipeline
p = Pipeline(stages=[vector_assembler, spark_model]).fit(df)
p.write().overwrite().save('lhorizontal_sparktorch')
```

```python
# Example of loading the pipeline
loaded_pipeline = PysparkPipelineWrapper.unwrap(PipelineModel.load('lhorizontal_sparktorch'))

# Run predictions and evaluation
predictions = loaded_pipeline.transform(df).persist()
evaluator1 = RegressionEvaluator(labelCol="Pressure", predictionCol="predictions", metricName="rmse")
evaluator2 = RegressionEvaluator(labelCol="Pressure", predictionCol="predictions", metricName="r2")

rmse = evaluator1.evaluate(predictions)
r2 = evaluator2.evaluate(predictions)

print("RMSE: %g" % rmse)
print("r2: %g" % r2)
```
